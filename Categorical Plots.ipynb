```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips=pd.read_csv('ACL_Export.csv')
```

```python
tips.head()
```

```python
tips[tips['total bills']>1900].head()
```

```python
tip2=tips[(tips['type'] == 'CN') | (tips['type'] =='IN')]
```

```python
tip2.head()
```

```python
import numpy as np
```

```python
#access the categorical plot based on some function or mean
#visualisation of a groupby action

sns.barplot(x='type',y='amount',data=tip2,estimator=np.std)
```

```python
sns.countplot(x='type',data=tips)
```

```python
#boxplot and volinplot is used to show the dist of quantitative data

sns.boxplot(x='day',y='amount',data=tip2,hue='type')
```

```python
sns.violinplot(x='day',y='amount',data=tips,hue='agree',split=True)
```

```python
sns.stripplot(x='day',y='amount',data=tips,jitter=True,hue='type')
```

```python
sns.violinplot(x='day',y='amount',data=tips)
sns.swarmplot(x='day',y='amount',data=tips,color='black',size=2)
```

```python
sns.catplot(x='day',y='amount',data=tips, kind='violin')
```
